```python
import pandas as pd

df = pd.read_csv('data/creditcard.csv')

# Get 5 actual fraud transactions
frauds = df[df['Class'] == 1].head(5)

print("REAL FRAUD TRANSACTIONS FROM DATASET:")
print("="*60)

for i, (idx, row) in enumerate(frauds.iterrows(), 1):
    print(f"\nFraud Example {i}:")
    print(f"Time: {row['Time']}")
    print(f"Amount: {row['Amount']}")
    print(f"V1: {row['V1']:.2f}")
    print(f"V2: {row['V2']:.2f}")
    print(f"V3: {row['V3']:.2f}")
    print(f"V4: {row['V4']:.2f}")
    print(f"V10: {row['V10']:.2f}")
    print(f"V12: {row['V12']:.2f}")
    print(f"V14: {row['V14']:.2f}")
    print(f"V17: {row['V17']:.2f}")
    print("-"*60)

# Get 2 legitimate transactions
legit = df[df['Class'] == 0].head(2)

print("\nLEGITIMATE TRANSACTIONS:")
print("="*60)

for i, (idx, row) in enumerate(legit.iterrows(), 1):
    print(f"\nLegit Example {i}:")
    print(f"Time: {row['Time']}")
    print(f"Amount: {row['Amount']}")
    print(f"V1: {row['V1']:.2f}")
    print(f"V2: {row['V2']:.2f}")
    print("-"*60)
```
